<font color='#008080'><b>Filtering Images by Extension</b></font>

```python
data_dir = '/kaggle/input/sad-and-happy-face-detection/data' 
image_exts = ['jpeg','jpg', 'bmp', 'png']
for image_class in os.listdir(data_dir): 
    for image in os.listdir(os.path.join(data_dir, image_class)):
        image_path = os.path.join(data_dir, image_class, image)
        try: 
            img = cv2.imread(image_path)
            tip = imghdr.what(image_path)
            if tip not in image_exts: 
                print('Image not in ext list {}'.format(image_path))
                os.remove(image_path)
        except Exception as e: 
            print('Issue with image {}'.format(image_path))
            # os.remove(image_path)
```

The above code snippet demonstrates how to filter images by their extensions in a given directory.

Here's how it works:

The variable <font color="#FF00FF">data_dir</font> stores the path to the directory where the images are located.
The list <font color="#FF00FF">image_exts</font> contains the desired image extensions, such as 'jpeg', 'jpg', 'bmp', and 'png'.
The code iterates over each image class (subdirectory) within the <font color="#FF00FF">data_dir</font> directory using <font color="#FF00FF">os.listdir(data_dir)</font>.
For each image class, it further iterates over each image within that class using <font color="#FF00FF">os.listdir(os.path.join(data_dir, image_class))</font>.
The <font color="#FF00FF">image_path</font> variable is created by joining the <font color="#FF00FF">data_dir</font>, <font color="#FF00FF">image_class</font>, and <font color="#FF00FF">image</font> names.
Inside the loop, the code attempts to read the image using <font color="#FF00FF">cv2.imread(image_path)</font>.
The <font color="#FF00FF">imghdr.what(image_path)</font> function is used to determine the image's format.
If the image's format (<font color="#FF00FF">tip</font>) is not found in the <font color="#FF00FF">image_exts</font> list, it indicates that the image is not in the desired extensions. In this case, a message is printed indicating that the image is not in the extension list, and the image file is removed using <font color="#FF00FF">os.remove(image_path)</font>.
If any exceptions occur during the image processing, an error message is printed, and the code has the option to remove the image file as well.
This code snippet helps ensure that only images with specified extensions are retained in the directory, allowing for cleaner and more focused data for further analysis or training.

<font color='#008080'><b>Sequential</b></font>: This module is used to initialize a sequential model, which is a linear stack of layers. It allows us to build a deep learning model by adding layers one by one.
<font color='#008080'><b>Conv2D</b></font>: This module is used to add a 2D convolutional layer to the model. It performs a convolution operation on the input data to extract relevant features.

<font color='#008080'><b>Creating a Sequential Model</b></font>

```python
model = Sequential()
```

The above code snippet demonstrates how to create a sequential model using the <font color='#008080'><b>Sequential</b></font> class from TensorFlow Keras.

Here's what it does:

The <font color='#008080'><b>Sequential</b></font> function is called to initialize an empty sequential model.

<font color='#008080'><b>Building the Convolutional Neural Network (CNN) Architecture</b></font>

```python
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

The provided code snippet demonstrates the process of building a Convolutional Neural Network (CNN) architecture using the Sequential model from TensorFlow Keras. The architecture consists of several convolutional, pooling, and dense layers.

Here's an explanation of each line of code:

<font color='#008080'><b>model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))</b></font>
This line adds a Conv2D layer to the model with 16 filters, a kernel size of (3,3), a stride of 1, and ReLU activation. It also specifies the input shape as (256,256,3), representing an input image with a height and width of 256 pixels and 3 color channels (RGB).

<font color='#008080'><b>model.add(MaxPooling2D())</b></font>
This line adds a MaxPooling2D layer to the model, which performs max pooling operation with default parameters.

<font color='#008080'><b>model.add(Conv2D(32, (3,3), 1, activation='relu'))</b></font>
This line adds another Conv2D layer with 32 filters, a kernel size of (3,3), a stride of 1, and ReLU activation.

<font color='#008080'><b>model.add(MaxPooling2D())</b></font>
This line adds another MaxPooling2D layer.

<font color='#008080'><b>model.add(Conv2D(16, (3,3), 1, activation='relu'))</b></font>
This line adds another Conv2D layer with 16 filters, a kernel size of (3,3), a stride of 1, and ReLU activation.

<font color='#008080'><b>model.add(MaxPooling2D())</b></font>
This line adds another MaxPooling2D layer.

<font color='#008080'><b>model.add(Flatten())</b></font>
This line adds a Flatten layer to convert the 2D feature maps into a 1D feature vector.

<font color='#008080'><b>model.add(Dense(256, activation='relu'))</b></font>
This line adds a Dense layer with 256 units and ReLU activation.

<font color='#008080'><b>model.add(Dense(1, activation='sigmoid'))</b></font>
This line adds the final Dense layer with 1 unit and sigmoid activation, representing the output layer for binary classification.

By stacking these layers in a sequential manner, you create a deep learning architecture capable of learning hierarchical features from input images and making predictions.

Here's what it does:
