<font color='#008080'><b>Filtering Images by Extension</b></font>

```python
data_dir = '/kaggle/input/sad-and-happy-face-detection/data' 
image_exts = ['jpeg','jpg', 'bmp', 'png']
for image_class in os.listdir(data_dir): 
    for image in os.listdir(os.path.join(data_dir, image_class)):
        image_path = os.path.join(data_dir, image_class, image)
        try: 
            img = cv2.imread(image_path)
            tip = imghdr.what(image_path)
            if tip not in image_exts: 
                print('Image not in ext list {}'.format(image_path))
                os.remove(image_path)
        except Exception as e: 
            print('Issue with image {}'.format(image_path))
            # os.remove(image_path)
```

The above code snippet demonstrates how to filter images by their extensions in a given directory.

Here's how it works:

The list <font color="#FF00FF">image_exts</font> contains the desired image extensions, such as 'jpeg', 'jpg', 'bmp', and 'png'.
The code iterates over each image class (subdirectory) within the <font color="#FF00FF">data_dir</font> directory using <font color="#FF00FF">os.listdir(data_dir)</font>.
For each image class, it further iterates over each image within that class using <font color="#FF00FF">os.listdir(os.path.join(data_dir, image_class))</font>.
The <font color="#FF00FF">image_path</font> variable is created by joining the <font color="#FF00FF">data_dir</font>, <font color="#FF00FF">image_class</font>, and <font color="#FF00FF">image</font> names.
Inside the loop, the code attempts to read the image using <font color="#FF00FF">cv2.imread(image_path)</font>.
The <font color="#FF00FF">imghdr.what(image_path)</font> function is used to determine the image's format.
If the image's format (<font color="#FF00FF">tip</font>) is not found in the <font color="#FF00FF">image_exts</font> list, it indicates that the image is not in the desired extensions. In this case, a message is printed indicating that the image is not in the extension list, and the image file is removed using <font color="#FF00FF">os.remove(image_path)</font>.

<font color='#008080'><b>Splitting the Data into Train, Validation, and Test Sets</b></font>

```python
train_size = int(len(data)*.7)
val_size = int(len(data)*.2)
test_size = int(len(data)*.1)
```

The above code snippet showcases how to split the data into train, validation, and test sets based on the specified proportions.

Here's how it works:

The length of the data is calculated using the <font color='#008080'><b>len()</b></font> function, representing the total number of samples in the dataset.
The train size is determined by multiplying the length of the data by 0.7, indicating that 70% of the data will be allocated for training.
The validation size is computed by multiplying the length of the data by 0.2, indicating that 20% of the data will be reserved for validation.
The test size is obtained by multiplying the length of the data by 0.1, indicating that 10% of the data will be used for testing.
By splitting the data into different sets, you can effectively evaluate the performance of your deep learning model on unseen data and fine-tune its parameters during the training process. The train set is used to optimize the model's parameters, the validation set helps in monitoring the model's performance and preventing overfitting, while the test set provides an unbiased assessment of the final model's accuracy.

<font color='#008080'><b>Splitting the Data into Train, Validation, and Test Sets</b></font>

```python
train = data.take(train_size)
val = data.skip(train_size).take(val_size)
test = data.skip(train_size+val_size).take(test_size)
```

The above code snippet demonstrates how to split the data into train, validation, and test sets using the TensorFlow Dataset API.

Here's how it works:

The <font color='#008080'><b>take()</b></font> method is used to create a new dataset containing the specified number of elements from the beginning of the original dataset.
The <font color='#008080'><b>skip()</b></font> method is used to create a new dataset that skips the specified number of elements from the beginning of the original dataset.
By combining these methods, we can split the data into train, validation, and test sets based on the given proportions.
The <font color='#008080'><b>train</b></font>, <font color='#008080'><b>val</b></font>, and <font color='#008080'><b>test</b></font> datasets can be used for training, validation, and testing of machine learning models respectively.

Splitting the data into different sets is an important step in machine learning to ensure proper evaluation and generalization of models.

<font color='#008080'><b>Importing Required Modules</b></font>

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D,MaxPooling2D , Dense , Flatten , Dropout
```

The above code snippet demonstrates how to import the necessary modules from the TensorFlow Keras library for building a convolutional neural network (CNN) model.

Here's what each module is used for:

<font color='#008080'><b>Sequential</b></font>: This module is used to initialize a sequential model, which is a linear stack of layers. It allows us to build a deep learning model by adding layers one by one.
<font color='#008080'><b>Conv2D</b></font>: This module is used to add a 2D convolutional layer to the model. It performs a convolution operation on the input data to extract relevant features.
<font color='#008080'><b>MaxPooling2D</b></font>: This module is used to add a 2D max-pooling layer to the model. It reduces the spatial dimensions of the input data by taking the maximum value within each pooling window.
<font color='#008080'><b>Dense</b></font>: This module is used to add a fully connected (dense) layer to the model. It connects every neuron in the previous layer to the neurons in the current layer.
<font color='#008080'><b>Flatten</b></font>: This module is used to flatten the input data into a 1D array, which is required before passing it to the dense layers.
<font color='#008080'><b>Dropout</b></font>: This module is used to add a dropout regularization layer to the model. It randomly drops a certain percentage of neurons during training to prevent overfitting.

These modules provide the necessary building blocks for constructing a CNN model, which is commonly used for image classification tasks in deep learning.

<font color='#008080'><b>Creating a Sequential Model</b></font>

```python
model = Sequential()
```

The above code snippet demonstrates how to create a sequential model using the <font color='#008080'><b>Sequential</b></font> class from TensorFlow Keras.

Here's what it does:

The <font color='#008080'><b>Sequential</b></font> function is called to initialize an empty sequential model.
This model acts as a container for stacking layers.
Once the model is created, you can add layers to it using the <font color='#008080'><b>add()</b></font> method.
By building a sequential model, you can easily add layers one after another to create a deep learning architecture. The sequential nature of the model allows for a straightforward representation of the flow of data through the layers.

```python
model.add(Conv2D(16, (3,3), 1, activation='relu', input_shape=(256,256,3)))
model.add(MaxPooling2D())
model.add(Conv2D(32, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Conv2D(16, (3,3), 1, activation='relu'))
model.add(MaxPooling2D())
model.add(Flatten())
model.add(Dense(256, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile('adam',loss = tf.losses.BinaryCrossentropy(), metrics=['accuracy'])
```

```python
model.summary()
```

```python
logdir = 'logs'
```

```python
tensorboard_callback = tf.keras.callbacks.TensorBoard(log_dir = logdir)
```

```python
hist = model.fit(train,epochs=30,validation_data=val,callbacks=[tensorboard_callback])
```

```python
fig = plt.figure()
plt.plot(hist.history['loss'], color = 'teal', label='loss')
plt.plot(hist.history['val_loss'],color='orange',label='val_loss')
fig.suptitle('Loss', fontsize=20)
plt.legend(loc="upper left")
plt.show()
```

```python
fig = plt.figure()
plt.plot(hist.history['accuracy'] , color ='teal' , label ='accuracy')
plt.plot(hist.history['val_accuracy'], color ='orange' , label = 'val_accuracy')
fig.suptitle('Accuracy' , fontsize =20)
plt.legend(loc="upper left")
plt.show()
```

```python
from tensorflow.keras.metrics import Precision, Recall, BinaryAccuracy
```

```python
pre = Precision()
re = Recall()
acc = BinaryAccuracy()
```

```python
for batch in test.as_numpy_iterator():
    X, y = batch
    yhat = model.predict(X)
    pre.update_state(y,yhat)
    re.update_state(y,yhat)
    acc.update_state(y,yhat)
```

```python
print(pre.result(), re.result(),acc.result())
```
